Q6

```python
flight=sns.load_dataset('flights')
```

```python
flight
```

```python
import matplotlib.pyplot as plt
heatmap_data=flight.pivot('month','year','passengers')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data,cmap='YlGnBu',annot=True,fmt='d',cbar=True)
plt.title('Passenger Count Heatmap (Flights Dataset)')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
